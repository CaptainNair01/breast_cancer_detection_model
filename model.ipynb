```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection and Processing

```python
#loading the data from sklearn
bc_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(bc_dataset)
```

```python
#load the data into the data frame
df = pd.DataFrame(bc_dataset.data, columns = bc_dataset.feature_names)

#printing the first 5 rows
df.head()
```

```python
#adding the 'target' column to the df
df['label'] = bc_dataset.target
```

```python
#print last 5 rows of df
df.tail()
```

```python
#number of rows and columns
df.shape
```

```python
#getting info about the data
df.info()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#statistical measures about the data
df.describe()
```

```python
#checking the distribution of the target variable
df['label'].value_counts()
```

```python
df.groupby('label').mean()
```

```python
#Seperating the features and target
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Testing and Traininng data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of my neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#compiling the neural network
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#training the neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=100)
```

```python
#visualizing the accuracy and the loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
#Accuracy of the model on the test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
